Consider the coherent state:

$$\ket{\alpha} = e^{-\frac{\mid\alpha\mid^2}{2}} \sum_{n=0}^{\infty} \frac{\alpha^n}{\sqrt{n!}} \ket{n}$$

It is an eigenstate of the annhilation operator $\hat{a}$ with eigenvalue $\alpha$:  $$\hat{a}\ket{\alpha} = \alpha\ket{\alpha}$$

**Demonstration:**
$$
\ket{\alpha} = \sum_{n}\ket{n}\bra{n}\ket{\alpha} = \sum_{n=0} \braket{n}{\alpha}\ket{n} = \sum_{n}c_{n}(\alpha)\ket{n} \\
\hat{a}\ket{\alpha} = \sum_{n}c_{n}(\alpha)\hat{a}\ket{n} = \sum_{n} c_{n}(\alpha) \sqrt{n} \ket{n-1} = \alpha \ket{\alpha} = \alpha\sum_{n}c_{n}(\alpha)\ket{n}
$$

We get:  
$$
\sqrt{n}c_n(\alpha) = \alpha c_{n-1}(\alpha) \\
c_{n}(\alpha) = \frac{\alpha}{\sqrt(n)}c_{n-1}(\alpha) = ... = \frac{\alpha^n}{\sqrt(n!)}c_{0}(\alpha)
$$

The state becomes: 
$$
\ket{\alpha} = \sum_{n}\frac{\alpha^n}{\sqrt(n!)}c_{0}(\alpha) \ket{n}
$$

Normalize it: 
$$
1 = \braket{\alpha}{\alpha} = |c_{0}(\alpha)|^2 \sum_{n,m} \frac{\alpha^{m^*}}{\sqrt{m!}} \frac{\alpha^n}{\sqrt{n!}} \braket{m}{n} = |c_{0}(\alpha)|^2 \sum_{n} \frac{|\alpha|^{2n}}{n!} =  |c_{0}(\alpha)|^2 e^{|\alpha|^2} \\
c_{0}(\alpha) = e^{\frac{-|\alpha|^2}{2}}
$$
In the end we have:
$$
\alpha = e^{\frac{-|\alpha|^2}{2}}\sum_{n}\frac{\alpha^n}{\sqrt(n!)} \ket{n}
$$
$\blacksquare$


For consitency we can show:
$$
\hat{a}\ket{\alpha} = \alpha\ket{\alpha} \\
\hat{a}\ket{\alpha} = e^{\frac{-|\alpha|^2}{2}}\sum_{n=0}^{\infty}\frac{\alpha^n}{\sqrt(n!)} \hat{a}\ket{n} = e^{\frac{-|\alpha|^2}{2}}\sum_{n=1}^{\infty}\frac{\alpha^n}{\sqrt((n-1)!)} \ket{n-1} = \alpha e^{\frac{-|\alpha|^2}{2}}\sum_{n=1}^{\infty}\frac{\alpha^{n-1}}{\sqrt((n-1)!)} \ket{n-1} = \alpha e^{\frac{-|\alpha|^2}{2}}\sum_{n=0}^{\infty}\frac{\alpha^n}{\sqrt(n!)} \ket{n} = \alpha\ket{\alpha}
$$

$$
1 = \braket{\alpha}{\alpha} = e^{-|\alpha|^2} \sum_{n,m} \frac{\alpha^{m^*}}{\sqrt{m!}} \frac{\alpha^n}{\sqrt{n!}} \braket{m}{n} = e^{-|\alpha|^2} \sum_{n} \frac{|\alpha|^{2n}}{n!} =  e^{-|\alpha|^2} e^{|\alpha|^2} = 1
$$



Now introduce the displacement operator:
$$
\hat{D}(\alpha) = \exp(\alpha\hat{a}^\dagger - \alpha^*\hat{a})
$$

We can show that:
$$
\ket{\alpha} = \hat{D}(\alpha)\ket{0}
$$

**Demonstration:**

First of all notice that the operator is unitary: $\ \ \ \hat{D}(\alpha)\hat{D}^\dagger(\alpha) = \hat{D}^\dagger(\alpha)\hat{D}(\alpha) = 1$

Recall this relation: if $[A,[A,B]] = 0$ and $[B,[A,B]] = 0$, then $ e^A e^B = e^{A+B}e^{\frac{1}{2}[A,B]}$

For the displacement operator, we got:
$$
[\alpha\hat{a}^\dagger,-\alpha^*\hat{a}] = -|\alpha|^2 = C \\
[\alpha\hat{a}^\dagger, C] =0 \ \ \ \ \ \ \ [-\alpha^*\hat{a}, C] = 0
$$

So we can rewrite the operator as:
$$
\hat{D}(\alpha) = \exp(\alpha\hat{a}^\dagger - \alpha^*\hat{a}) = e^{-\frac{|\alpha|^2}{2}} \exp(\alpha\hat{a}^\dagger) \exp(-\alpha^*\hat{a})
$$

Now apply to the vacuum state $\ket{0}$:
$$
\hat{D}(\alpha)\ket{0} = e^{-\frac{|\alpha|^2}{2}} \exp(\alpha\hat{a}^\dagger) \exp(-\alpha^*\hat{a}) \ket{0}
$$

-  $exp(-\alpha^*\hat{a})\ket{0} = (1 - \alpha^*\hat{a} + \frac{\alpha^{*^2}}{2!}\hat{a}^2 + ...)\ket{0} = \ket{0}$
-  $exp(\alpha\hat{a}^\dagger)\ket{0} = (1 + \alpha\hat{a}^\dagger + \frac{\alpha^2}{2!}\hat{a}^{\dagger^2} + ...)\ket{0} = \sum_{n}\frac{(\alpha\hat{a}^\dagger)^n}{n!})\ket{0} = \sum_{n} \frac{\alpha^n}{\sqrt{n!}}\ket{n}$

We get:
$$
\hat{D}(\alpha)\ket{0} = e^{-\frac{|\alpha|^2}{2}} \sum_{n} \frac{\alpha^n}{\sqrt{n!}}\ket{n} = \ket{\alpha}
$$
$\blacksquare$

## Question 2

Prove:
$$
\hat{D}^{-1}(\alpha) \hat{a} \hat{D}(\alpha) = \hat{a} + \alpha
$$

**Demonstration:**

Recall the displacement operator and its self-adjoint (which coincides with the inverse, unitary):
$$
\hat{D}(\alpha) = \exp(\alpha\hat{a}^\dagger - \alpha^*\hat{a}) \\
\hat{D}^\dagger(\alpha) = \exp(\alpha^*\hat{a} - \alpha\hat{a}^\dagger) = \hat{D}^{-1}(\alpha)
$$

Recall the identity for a given operator B and a transformation $T=e^S$:
$$
e^S B e^{-S} = B + [S,B] + \frac{1}{2} [S,[S,B]] + ...
$$

So we obtain that our B is $\hat{a}$ and our S is $\alpha^*\hat{a} - \alpha\hat{a}^\dagger$:

- $[\alpha^*\hat{a} - \alpha\hat{a}^\dagger, \hat{a}] = -\alpha[\hat{a}^\dagger,\hat{a}] = \alpha$
- $[\alpha^*\hat{a} - \alpha\hat{a}^\dagger, \alpha] = 0$
- higher orders give zero contribution

In the end:
$$
\hat{D}^{-1}(\alpha) \hat{a} \hat{D}(\alpha) = \hat{a} + \alpha
$$

$\blacksquare$


Now we want to use this relation to prove:
$$
\overline{n} = |\alpha|^2
$$

**Demonstration:**


We have found that $\ \ \ket{\alpha} = \hat{D}(\alpha)\ket{0}$, now calculate $\overline{n}$:
$$
\overline{n} = \bra{\alpha} \hat{N} \ket{\alpha}
$$

where $\hat{N} = \hat{a}^\dagger \hat{a}$ is the photon number operator.

$$
\overline{n} = \bra{\alpha} \hat{N} \ket{\alpha} = \bra{\alpha} \hat{a}^\dagger\hat{a} \ket{\alpha} = \bra{0} \hat{D}^\dagger(\alpha) \hat{a}^\dagger\hat{a} \hat{D}(\alpha)\ket{0} = \bra{0} \hat{D}^\dagger(\alpha) \hat{a}^\dagger \hat{D}(\alpha) \ \ \ \hat{D}^\dagger(\alpha)  \hat{a} \hat{D}(\alpha)\ket{0}
$$

Now we know that:
$$
\hat{D}^{-1}(\alpha) \hat{a} \hat{D}(\alpha) = \hat{D}^\dagger(\alpha) \hat{a} \hat{D}(\alpha) = \hat{a} + \alpha
$$

With the same argument as before we obtain:
$$
\hat{D}^\dagger(\alpha) \hat{a}^\dagger \hat{D}(\alpha) = \hat{a}^\dagger + \alpha^*
$$

So we get:
$$
\overline{n} = \bra{\alpha} \hat{N} \ket{\alpha} = \bra{0} (\hat{a}^\dagger + \alpha^*)(\hat{a} + \alpha) \ket{0} = \bra{0} (\hat{N} + |\alpha|^2 + \hat{a}^\dagger \alpha + \alpha^*\hat{a}) \ket{0} = |\alpha|^2
$$

In the end:
$$
\overline{n} = |\alpha|^2
$$

$\blacksquare$

Consider the total Hamiltonian:
$$
\hat{H} = \hat{H}_{atom} + \hat{H}_{light} + \hat{H}_{atom-light}

Change of frame: transormation to rotating frame (interaction picture)
$$
R = R_{atom} \otimes R_{light}
$$

-  $R_{atom} = e^{i\frac{\hat{H}_{atom}}{\hbar} t} = e^{i\omega_{eg}\ket{e}\bra{e} \ t}$
-  $R_{light} = e^{i\frac{\hat{H}_{light}}{\hbar} t} = e^{i\omega_c \hat{a}^\dagger \hat{a} \ t}$

The transformation of the Hamiltonian follows the rule:
$$
\hat{H}' = R\hat{H}R^\dagger - i\hbar R \dot{R}^\dagger
$$

Analyze $\hat{H}_{atom}$, the transformation operator $R_{light}$ does not influence $H_{atom}$ (they commute):

-  $R_{atom}\hat{H}_{atom}R^\dagger_{atom} = \hat{H}_{atom}$

Analyze $\hat{H}_{light}$, the transformation operator $R_{atom}$ does not influence $H_{light}$ (they commute):

-  $R_{light}\hat{H}_{light}R^\dagger_{light} = \hat{H}_{light}$
-  $- i\hbar R_{light} \dot{R}^\dagger_{light} = -i\hbar e^{i\omega_c \hat{a}^\dagger \hat{a} \ t} (-i\omega_c) e^{-i\omega_c \hat{a}^\dagger \hat{a} \ t} \hat{a}^{\dagger}\hat{a} = -\hat{H}_{light}$


So we get: 

$\hat{H}' = R \hat{H}_{atom-light} R^\dagger = \hat{H}'_{atom-light}$

Analyze $\hat{H}_{atom-light}$, the transformation operator $R_{atom}$ does not influence $H_{atom-light}$ because the latter acts only on photons space (so they commute):

-   $R_{light}\hat{a}R^\dagger_{light} = e^{i\omega_{eg} \hat{a}^\dagger\hat{a} \ t} \ \hat{a} \ e^{-i\omega_{eg} \hat{a}^\dagger\hat{a} \ t} = e^{i\frac{\hat{H}_{light}}{\hbar} t} \ \hat{a} \ e^{-i\frac{\hat{H}_{light}}{\hbar} t} = \hat{a}(t) = \hat{a}e^{-i\omega_c \ t}$
-   $R_{light}\hat{a}^\dagger R^\dagger_{light} = e^{i\omega_{eg} \hat{a}^\dagger\hat{a} \ t} \ \hat{a}^\dagger \ e^{-i\omega_{eg} \hat{a}^\dagger\hat{a} \ t} = e^{i\frac{\hat{H}_{light}}{\hbar} t} \ \hat{a}^\dagger \ e^{-i\frac{\hat{H}_{light}}{\hbar} t} = \hat{a}_\dagger(t) = \hat{a}^\dagger e^{i\omega_c \ t}$

We end up with:
$$
\hat{H}' = -d_0 E_0 cos(\omega t) (\hat{a} e^{-i\omega_c t} - \hat{a}^\dagger e^{i\omega_c t}) = -\frac{d_0}{2} E_0 (e^{i\omega t} + e^{-i\omega t}) (\hat{a} e^{-i\omega_c t} - \hat{a}^\dagger e^{i\omega_c t})
$$

Now apply the Rotating Wave approximation, neglect terms with fast oscillations frequency (average zero in period of slow oscillations frequency) $|\omega - \omega_c |\ll\omega+\omega_c $, and use the fact that we are in perfect resonance $\omega = \omega_c$:
$$
\hat{H}' = -\frac{d_0}{2} E_0 \ (\hat{a} e^{i( \omega - \omega_c )t} - \hat{a}^\dagger e^{-i( \omega - \omega_c )t}) = -\frac{d_0}{2} E_0 (\hat{a} - \hat{a}^\dagger)
$$

Now compute the time evolution operator (assuming $\ket{\psi(t_0 = 0) = \ket{0}}$) and compare it to the displacement operator :
$$
\ket{\psi(t)} = U(t,t_0)\ket{\psi(t_0)} = e^{-i\frac{H'}{\hbar} t}\ket{0} \\
\ket{\psi(t)} = e^{i\frac{d_0 E_0}{2\hbar} (\hat{a} - \hat{a}^\dagger) t}\ket{0} = e^{-i\frac{d_0 E_0}{2\hbar} (\hat{a}^\dagger - \hat{a}) t}\ket{0}
$$

so we get: $$U(t,t_0) = \exp\left(-i\frac{d_0 E_0}{2\hbar} (\hat{a}^\dagger - \hat{a}) t\right)$$



while the displacement operator is: $$\hat{D}(\alpha) = \exp\left(\alpha\hat{a}^\dagger - \alpha^*\hat{a}\right)$$

So to get the match we have to consider: $$\alpha = -i \frac{d_0 E_0}{2\hbar} t$$

Consider $\hbar = 1$ and compute the time evolution of the vacuum state $\ket{0}$ and the match with the coherent state $\ket{\alpha}$

```python
import numpy as np
import scipy as sp
from scipy import linalg
import matplotlib
from matplotlib import pyplot as plt
from scipy.special import factorial
from scipy import optimize
from scipy.signal import find_peaks
from scipy.integrate import odeint
```

```python
d0 = 1
E0 = 1j
n = 100 # Hilbert space 100 dim (100 possibile photons states)

# Annihilation operator
def a(n):
    a = np.zeros((n,n))
    for i in range(n):
        for j in range(i,n):
            if j == i+1: a[i,j] = np.sqrt(j)
    return a

# Creation operator
def a_d(n):
    a_d = np.zeros((n,n))
    for i in range(n):
        for j in range(i,n):
            if j == i+1: a_d[j,i] = np.sqrt(j)
    return a_d

# Hamiltonian atom-light in interaction picture
def H_AL(d0, E0, n):
    ham = - (d0/2) * E0 * (a(n) - a_d(n))
    return ham

# Time evolution operator
def U(time, d0, E0, n):
    U = sp.linalg.expm(-1j * time * H_AL(d0, E0, n))
    return U
```

```python
def state_evolution(time,psi0):
    state = np.dot(U(time, d0, E0, n), psi0)
    return state
```

```python
t_start = 0
t_stop = 10
dt = 0.01
time_line = np.arange(t_start,t_stop,dt)
```

```python
# Define initial state: vacuum state
psi0 = np.concatenate((np.array([1.]),np.zeros(n-1)),axis=0)
```

```python
# Wave function of the evolved state

psit = [state_evolution(t,psi0) for t in time_line]

plt.plot(time_line, psit)
plt.xlabel('Time')
plt.ylabel('Wave function')
plt.show()
```

Now select a photon state $n$, calculate its probability of occuring and compare it to the one given by the coherent state definition (so by the dispacement operator):

```python
n_state = []
for i in psit:
        n_state.append(i[3])
p_n = np.abs(n_state)**2
```

```python
def poisson(d0, E0, n, t):
    alpha = -1j * (d0/2) * E0 * t
    poisson = np.exp(-np.abs(alpha)**2) * alpha**(2*n) / factorial(n)
    return poisson
```

```python
plt.plot(time_line, p_n, label='Probability of photon state n')
plt.plot(time_line,poisson(d0,E0,3,time_line), label='Poisson distribution for photon state n')
plt.legend(loc='best')
plt.show()
```

This is the result with all the states:

The probabilities coincide. The displacement operator acts on vacuum state as the time evolution operator of the system in the interaction picture.

Now find the collapse and revival of the superfluid parameter.

```python
U_energy = 1
n = 100 #100  #100
N_atoms = 50  #10  #30
```

```python
# Annihilation operator
def a(n):
    a = np.zeros((n,n))
    for i in range(n):
        for j in range(i,n):
            if j == i+1: a[i,j] = np.sqrt(j)
    return a


# Creation operator
def a_d(n):
    a_d = np.zeros((n,n))
    for i in range(n):
        for j in range(i,n):
            if j == i+1: a_d[j,i] = np.sqrt(j)
    return a_d


# Number operator
def n_op(n):
    #n_op = np.diag(np.sqrt(np.arange(n)))
    n_op = np.dot(a_d(n),a(n))
    return n_op

# Hamiltonian atom-light in interaction picture
def H_BH(U, n):
    h_bh = (U/2) * np.dot(n_op(n),(n_op(n) - np.identity(n)))
    return h_bh

# Time evolution operator
def U_BH(time, U, n):
    U_bh = sp.linalg.expm(-1j * time * H_BH(U, n))
    return U_bh
```

```python
# Time line
t_start = 0
t_stop = 20
dt = 0.01
time_line_bc = np.arange(t_start,t_stop,dt)
```

```python
def alfa(n,N_atoms):
    alfa = np.zeros(n)
    for m in range(n):
        alfa[m] = np.exp(-N_atoms/2) * np.sqrt(N_atoms)**m/(np.sqrt(factorial(m)))
    return alfa
```

```python
def U_BH_conj(time, U, n):
    U_bh = sp.linalg.expm(1j * time * H_BH(U, n))
    return U_bh
```

```python
psi_time = []
for i in time_line_bc:
    psi_time.append(np.dot(U_BH_conj(i, U_energy, n),np.dot(a(n), U_BH(i, U_energy, n))))
psi_time = np.array(psi_time)
```

```python
psi_time.shape
```

```python
alfa_time0 = alfa(n,N_atoms)
coeff = []
for i in range(len(psi_time)):
    sup_fluid = np.dot(alfa_time0, np.dot(psi_time[i,:,:],alfa_time0))
    coeff.append(sup_fluid)
coeff = np.array(coeff)
```

```python
plt.plot(time_line_bc,coeff)
#plt.xlim(0,10)
```

```python
time_restrict = time_line_bc[time_line_bc < 20]
coeff_restrict = coeff[time_line_bc < 20]
```

```python
plt.plot(time_restrict,coeff_restrict)
plt.xlim(0,7)
```

We can clearly see the trend of collapse and revival.
